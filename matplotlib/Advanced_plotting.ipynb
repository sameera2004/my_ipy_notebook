# Adavnced plotting in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl

# Easily switch between interactive and static matplotlib plots
interactive_mode = False

if interactive_mode:
    %matplotlib notebook
else:
    %matplotlib inline

backend = mpl.get_backend()
```

#####This method involves storing various elements of the of the plots in variables (these are objects in object-oriented terminology). 

#### Allowing users to easily handle multiple figures/axes without getting confused as to which one is currently active.

```python
fig = plt.figure()  # create a figure object
ax = fig.add_subplot(1, 1, 1)  # create an axes object in the figure
ax.plot([1, 2, 3, 4])
ax.set_ylabel('some numbers')
plt.show()
```

```python
fig1 = plt.figure()
fig2 = plt.figure()
ax1 = fig1.add_subplot(1, 1, 1)
ax2 = fig2.add_subplot(2, 1, 1)
ax3 = fig2.add_subplot(2, 1, 2)
```

## Twin axes

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()
# similarly, twiny would return a second set of axes sharing the y-axis, but with a separate x-axis
t = np.linspace(0., 10., 100)
ax1.plot(t, t ** 2, 'b-')
ax2.plot(t, 1000 / (t + 1), 'r-')
ax1.set_ylabel('Density (cgs)', color='red')
ax2.set_ylabel('Temperature (K)', color='blue')
ax1.set_xlabel('Time (s)')
plt.show()
```

##Controlling the appearance of plots

```python
# Initialize the figure and subplot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Set the font size via a keyword argument
ax.set_title("New Plot", fontsize='large')

# Retrieve an element of a plot and set properties
for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize('x-large')
    tick.set_fontname('Times New Roman')
    tick.set_color('red')
    tick.set_weight('extra bold')   
fig.canvas.draw()     
```

```python

plt.rc('xtick', color='r', labelsize='medium', direction='out')
plt.rc('xtick.major', size=4, pad=4)
plt.rc('xtick.minor', size=2, pad=4)
```

### Adding a legend

```python


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(1., 8., 30)
ax.plot(x, x ** 1.5, 'ro', label='density')
ax.plot(x, 20/x, 'bx', label='temperature')
ax.legend()
```

## Adding a colorbar

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
image = np.random.poisson(10., (100, 80))
i = ax.imshow(image, interpolation='nearest', cmap='Dark2')
fig.colorbar(i)  # note that colorbar is a method of the figure, not the axes
plt.show()
```

#### If you want to customize exactly where the colorbar appears, you can define a set of axes,
####and pass it to colorbar via the cax= argument:
#### To fix the imshow automatically modifies the axes to make that pixels are square. 
####We can fix this with aspect='auto':

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.8])
image = np.random.poisson(10., (100, 80))
i = ax.imshow(image, aspect='auto', interpolation='nearest')
colorbar_ax = fig.add_axes([0.7, 0.1, 0.05, 0.8])
fig.colorbar(i, cax=colorbar_ax)
plt.show()
```

### Use the colorbar in other inplots

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.6, 0.8])
x = np.random.random(400)
y = np.random.random(400)
c = np.random.poisson(10., 400)
s = ax.scatter(x, y, c=c, edgecolor='none')
ax.set_xlim(0., 1.)
ax.set_ylim(0., 1.)
colorbar_ax = fig.add_axes([0.7, 0.1, 0.05, 0.8])
fig.colorbar(s, cax=colorbar_ax)
plt.show()
```

##  Custom ticks and labels

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(0, 6)
ax.plot(x, np.cos(x), 'o')
ax.set_xticklabels(["A", "B", "M", "O", "5", "6"])
ax.set_yticklabels([2, 6, 7, 9, 10, 15])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([0.1, 0.5, 0.7])
ax.set_xticklabels(['a', 'b', 'c'])
ax.set_yticks([0.2, 0.4, 0.8])
ax.set_yticklabels(['first', 'second', 'third'])
plt.show()
```
