```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
image = np.random.random((100,100))
```

#  Set axes ticks to differnt ticks(labels) names

```python
fig,axes = plt.subplots()
axes.imshow(image, vmax=1e0)
axes.set_title("Random Data")
plt.show()
```

```python
fig,axes = plt.subplots()
axes.imshow(image, vmax=1e0)
axes.set_title("Random Data")
axes.set_xticks([0, 20, 40, 60, 80])
axes.set_xticklabels(["a", "b", "c", "d", "e"])
plt.show()
```

```python
fig,axes = plt.subplots()
axes.imshow(image, vmax=1e0)
axes.set_title("Random Data")
axes.set_xticks([0, 20, 40, 60, 80])
axes.set_xticklabels(["a", "b", "c", "d", "e"])
axes.set_yticks([0, 20, 40, 60, 80])
axes.set_yticklabels([2, 3, 4, 5, 6])
plt.show()
```
