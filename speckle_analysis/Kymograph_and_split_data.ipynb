```python
def split_the_data(splits, dataset):
    """Helper function to split the data into different groupings
    
    Parameters
    ----------
    splits : list
        List of splits including lower and upper bounds
    dataset : sliceable
        The dataset to split at the points defined by `splits`
    
    Returns
    -------
    data : list
        List of data split at the boundaries defined in `splits`
    ranges : list
        List of labels for the data in a format like: ['0:300', '300:500']
    """
    data = []
    ranges = []
    for lower, upper in zip(splits, splits[1:]):
        data.append(dataset[lower:upper])
        ranges.append('%s:%s' % (lower, upper))
    return data, ranges
```

```python
data_dir = "/Volumes/Data/BeamLines/CHX/Luxi_description_files_for_duke/Duke_data/"
duke_ndata = np.load(data_dir+"duke_img_1_5000.npy")
duke_dark = np.load(data_dir+"duke_dark.npy")

duke_data = []
for i in range(duke_ndata.shape[0]):
    duke_data.append(duke_ndata[i] - duke_dark)
    
duke_data=np.asarray(duke_data)
```

```python
# fake that this is 5 different points in the same sample
splits = [0, 500, 1500, 2000, 3500, duke_ndata.shape[0]]
split_data, ranges = split_the_data(splits, duke_ndata)
data_labels = ['point_%s' % (i+1) for i, region in zip(range(len(split_data)), ranges)]
data_dict = {label: data for label, data in zip(data_labels, split_data)}

# subtract the dark image from the dataset
```

```python
# subtract the dark image from the dataset
data_dict = {label: data - duke_dark for label, data in data_dict.items()}

mask1 = ~np.load(data_dir+"new_mask4.npy")
mask2 = ~np.load(data_dir+"Luxi_duke_mask.npy")

mask = mask1 + mask2

masked_data = {label: data * mask for label, data in data_dict.items()}
```

```python
# define the ROI parameters
starting_radius = 26
ring_width = 8
space_between_rings = 1
num_rings = 4

pixel_size = (0.1, 0.1)  # in mm
image_center = (133, 143) # in pixels

# get the inner/outer radii of the rings
ring_edges = roi.ring_edges(starting_radius, ring_width, space_between_rings, num_rings)
print('ring_edges are defined as [inner_radius, outer_radius]:\n%s' % ring_edges)
ring_centers = np.average(ring_edges, axis=1)
print('\nring_centers: %s' % ring_centers)

# rings is a labeled array where 0 is background and 1-4 are ROIS 1-4.
rings = roi.rings(edges=ring_edges, center=image_center, shape=(256,256))

# multiplying the rings with the image mask removes the bad information (beam stop, beam stop scatter, 
# bad pixels, etc..) from the rings ROI. See the plot below
rings_mask = rings*mask
```

```python
roi_kymograph = roi.roi_kymograph(split_data[0][200:1000], rings_mask, num=2)
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_xlabel('Pixel')
ax.set_ylabel('Frame')
ax.set_title('Kymograph')


im, cb = xsvs_plot.kymograph(ax, roi_kymograph, cmap='gist_earth', 
                         interpolation='None', aspect='auto', fps=100, frame_offset=800, clim=(0, 8))
```
