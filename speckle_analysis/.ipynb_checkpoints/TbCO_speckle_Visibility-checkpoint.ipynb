```python
import skxray.core.speckle as xsvs
import skxray.core.roi as roi
import skxray.core.correlation as corr
import skxray.core.utils as core

import xray_vision.mpl_plotting as mpl_plot
import xray_vision.mpl_plotting.speckle as speckle_plot

import numpy as np
import os, sys

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1.axes_grid import ImageGrid

interactive_mode = False

if interactive_mode:
    %matplotlib notebook
else:
    %matplotlib inline

backend = mpl.get_backend()
```

```python
data_dir = "/Volumes/Data/BeamLines/CSX_Data/X_data/"

TbCo_speckle = np.load(data_dir+"TbCo_speckle_visibility.npy")
TbCo_mask = np.load(data_dir+"TbCo_mask.npy")
```

```python
TbCo_speckle.shape
```

```python
plt.figure()
plt.imshow(TbCo_speckle)
plt.show()
```

```python
inner_radius = 35
width = 1
spacing = 0
num_rings = 1

#  create edges of the rings using roi.ring_edges
edges = roi.ring_edges(inner_radius, width, spacing, num_rings)
edges
```

```python
est_center = (113, 113)

rings = roi.rings(edges, est_center, TbCo_speckle.shape)
mask = ~TbCo_mask
rings_mask = rings*mask
```

```python
plt.figure()
plt.imshow(rings_mask)
plt.show()
```

```python
max_cts = roi.roi_max_counts((TbCo_speckle, ), rings_mask)
max_cts
```

```python
roi_pixel_values, index = roi.roi_pixel_values(TbCo_speckle, rings_mask)
```

```python
plt.figure()
plt.plot(roi_pixel_values[0])
plt.show()
```

```python
len(roi_pixel_values[0])
```

```python
images = TbCo_speckle.reshape(1, 230, 230)

spe_cts_all, std_dev = xsvs.xsvs((images, ), rings_mask, timebin_num=2,
                             number_of_img=8, max_cts=max_cts)
```

```python
spe_cts_all.shape
```

```python
plt.figure()
plt.plot(spe_cts_all[0,0])
plt.xlim(0, 4000)
```
