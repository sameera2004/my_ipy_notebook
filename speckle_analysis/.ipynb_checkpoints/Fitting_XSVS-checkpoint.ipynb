```python
data_dir = "/Volumes/Data/BeamLines/CHX/Luxi_description_files_for_duke/Duke_data/"
duke_ndata = np.load(data_dir+"duke_img_1_5000.npy")
duke_dark = np.load(data_dir+"duke_dark.npy")

duke_data = []
for i in range(duke_ndata.shape[0]):
    duke_data.append(duke_ndata[i] - duke_dark)
    
duke_data=np.asarray(duke_data)
```

```python
avg_img = np.average(duke_data, axis=0)

mask4 = np.load(data_dir+"new_mask4.npy")
mask = np.load(data_dir+"Luxi_duke_mask.npy")

# mask can be multiplied or added in the follwoing way
N_mask = (~mask * ~mask4)

fig = plt.figure(figsize=(15, 7))
grid = ImageGrid(fig, (1,1,1), nrows_ncols=(1, 2), share_all=True, 
                 cbar_mode='each', axes_pad=.3)
cmap = 'jet'
im2 = grid[1].imshow(np.log(N_mask*avg_img), vmax=1e0, cmap=cmap)
im1 = grid[0].imshow(np.log(avg_img), cmap=cmap)

grid[0].set_title("Duke Silica -- log of average")
grid[1].set_title("Duke Silica -- log of masked average")

grid.cbar_axes[0].colorbar(im1)
grid.cbar_axes[1].colorbar(im2)
# the following code is needed to make it easy to switch 
# between '%matplotlib inline' and '%matplotlib notebook'
plt.show()
```

#  Calling xsvs module for x-ray speckle visibility

```python
spe_cts_all, std_dev = xsvs.xsvs(images_sets, rings_mask, timebin_num=2,
                             number_of_img=8, max_cts=24)
```

# see the mean intensity of the images

```python
mean_int_sets, index_list = roi.mean_intensity_sets(images_sets, rings_mask)
combine_intensity = roi.combine_mean_intensity(mean_int_sets, index_list)

plt.plot(combine_intensity)
plt.show()

mean_int_ring = np.mean(combine_intensity, axis=0)
mean_int_ring
```

```python
num_times = 6
num_rois=num_rings
norm_bin_edges, norm_bin_centers = xsvs.normalize_bin_edges(num_times,
                                                            num_rois,
                                                            mean_int_ring, max_cts=24)
```

```python
fig, axes = plt.subplots(4, figsize=(8, 8))
for i in range(3):
    axes[0].set_xlabel("K/<K>")
    for j in range(4):
        art, = axes[j].plot(norm_bin_edges[i, j][:-1], spe_cts_all[i, j])
        axes[j].set_xlim(0, 4)
plt.show()
```

# Negative Binomial (Poisson-Gamma) distribution function

$P(K) = \frac{\Gamma(K + M)} {\Gamma(K + 1)\Gamma(M)}(\frac {M} {M + <K>})^M (\frac {<K>}{M + <K>})^K$

```python
from lmfit import minimize, Parameters, Model

binom_mod = Model(xsvs_fit.negative_binom_distribution)
```

# Poission Distribution

$ P(K) = \frac{<K>^K}{K!}\exp(-K)$

```python
poi_mod = Model(xsvs_fit.poisson_distribution)
```

# Gamma Distribution

$ P(K) =(\frac{M}{<K>})^M \frac{K^(M-1)}{\Gamma(M)}\exp(-M\frac{K}{<K>})$

```python
gamma_mod = Model(xsvs_fit.gamma_distribution)
```

```python
# 
```

```python
out1_bi = binom_mod.eval(bin_edges=norm_bin_edges[0, 3][:-1], K=1.023,
                      M=1.0234)

result_bi = binom_mod.fit(out1_bi, bin_edges=norm_bin_edges[0, 3][:-1],
                       K=1.023, M=1.024)

out1_po = poi_mod.eval(bin_edges=norm_bin_edges[0, 3][:-1], K=0.7896)

result_po = poi_mod.fit(out1_po, bin_edges=norm_bin_edges[0, 3][:-1], K=0.789)

out1_ga = gamma_mod.eval(bin_edges=norm_bin_edges[0, 3][:-1], K=0.78540586, M=1.234)

result_ga = gamma_mod.fit(out1_ga, bin_edges=norm_bin_edges[0, 3][:-1],
                          K=0.78540586, M=1.2234)


plt.figure()
plt.plot(norm_bin_edges[0, 3][:-1], spe_cts_all[0,3], 'ro', label='data')
plt.plot(norm_bin_edges[0, 3][:-1], result_bi.best_fit , '-b', label='negative_bi')
plt.plot(norm_bin_edges[0, 3][:-1], result_po.best_fit, '-go', label='poission' )
plt.plot(norm_bin_edges[0, 3][:-1], result_ga.best_fit, '--yo', label='gamma')
plt.ylim(-0.10, 0.50)
plt.legend()
plt.xlabel("K/<K>")
plt.ylabel("P(K)")
#plt.xlim(0, 4.0)
plt.show()
```

```python
out1_bi1 = binom_mod.eval(bin_edges=norm_bin_edges[1, 3][:-1], K=1.87,
                      M=2.3456)

result_bi1 = binom_mod.fit(out1_bi1, bin_edges=norm_bin_edges[1, 3][:-1],
                       K=1.56, M=2.345)

out1_po1 = poi_mod.eval(bin_edges=norm_bin_edges[1, 3][:-1], K=0.78540586*2)

result_po1 = poi_mod.fit(out1_po1, bin_edges=norm_bin_edges[1, 3][:-1], K=0.78540586*2)

out1_ga1 = gamma_mod.eval(bin_edges=norm_bin_edges[1, 3][:-1], K=0.78540586*2, M=1.51)

result_ga1 = gamma_mod.fit(out1_ga1, bin_edges=norm_bin_edges[1, 3][:-1],
                          K=0.78540586*2, M=1.51)


plt.figure()
plt.plot(norm_bin_edges[1, 3][:-1], spe_cts_all[1,3], 'ro', label='data')
plt.plot(norm_bin_edges[1, 3][:-1], result_bi1.best_fit , '-b', label='negative_bi')
#plt.plot(norm_bin_edges[1, 3][:-1], result_po1.best_fit, '-go', label='poission' )
plt.plot(norm_bin_edges[1, 3][:-1], result_ga1.best_fit, '--yo', label='gamma')
plt.ylim(-0.10, 0.55)
plt.legend()
plt.xlabel("K/<K>")
plt.ylabel("P(K)")
#plt.xlim(0, 4.0)
plt.show()
```

```python
out1_bi2 = binom_mod.eval(bin_edges=norm_bin_edges[2, 3][:-1], K=0.78540586*4,
                      M=2.23)

result_bi2 = binom_mod.fit(out1_bi2, bin_edges=norm_bin_edges[2, 3][:-1],
                       K=0.78540586*4, M=2.234)

out1_po2 = poi_mod.eval(bin_edges=norm_bin_edges[2, 3][:-1], K=4.123)

result_po2 = poi_mod.fit(out1_po2, bin_edges=norm_bin_edges[2, 3][:-1], K=4.12)

out1_ga2 = gamma_mod.eval(bin_edges=norm_bin_edges[2, 3][:-1], K=0.78540586*4, M=2.34)

result_ga2 = gamma_mod.fit(out1_ga2, bin_edges=norm_bin_edges[2, 3][:-1],
                          K=0.78540586*4, M=2.34)


plt.figure()
plt.plot(norm_bin_edges[2, 3][:-1], spe_cts_all[2,3], 'ro', label='data')
plt.plot(norm_bin_edges[2, 3][:-1], result_bi2.best_fit , '-b', label='negative_bi')
plt.plot(norm_bin_edges[2, 3][:-1], result_po2.best_fit, '-go', label='poission' )
plt.plot(norm_bin_edges[2, 3][:-1], result_ga2.best_fit, '--yo', label='gamma')
plt.ylim(-0.10, 0.35)
plt.legend()
plt.xlabel("K/<K>")
plt.ylabel("P(K)")
#plt.xlim(0, 3.0)
plt.show()
```
