```python
import skxray.core.speckle as xsvs
import skxray.core.roi as roi
import skxray.core.correlation as corr
import skxray.core.utils as core

import xray_vision.mpl_plotting as mpl_plot
import xray_vision.mpl_plotting.speckle as speckle_plot

import numpy as np
import os, sys

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1.axes_grid import ImageGrid

# Easily switch between interactive and static matplotlib plots
interactive_mode = False

if interactive_mode:
    %matplotlib notebook
else:
    %matplotlib inline

backend = mpl.get_backend()
```

##  get the data

```python
data_dir = "/Volumes/Data/BeamLines/CSX_Data/X_data/"

X_data = np.load(data_dir+"Sample_data_Xiao.npy")

X_data.shape
```

# Create a ring mask

```python
roi_data1 = np.array(([15, 10, 5, 5], (15, 15, 5, 5)), dtype=np.int64)
label_array1 = roi.rectangles(roi_data1, shape=X_data[0].shape)

plt.figure()
plt.imshow(label_array1)
plt.colorbar()
plt.show()
```

```python
def test_demo(ax, label_array, img_dim, image):
    """
    This will plot the reqiured roi's on the image
    """
    vmin, vmax = np.percentile(image, [100, 100])
    im = ax.imshow(image, interpolation='none', norm=LogNorm(), cmap='gray')
    im = ax.imshow(label_array, interpolation='none', norm=LogNorm())
    

fig, ax = plt.subplots()
plt.title("Data with ROI's")
test_demo(ax, label_array1, X_data[0].shape ,  X_data[0])
plt.show()    
    
```

#  Get the mean intensity for each ROI in the image list

```python
mean_int_sets, index_list = roi.mean_intensity(X_data, label_array1)
```

```python
plt.figure()
for i in range(2):
    plt.plot(mean_int_sets[:,i], label="ROI "+str(i+1))
plt.title("Mean intensities for each roi")
plt.legend()
plt.show()
```

```python
mean_int_ring = np.mean(mean_int_sets, axis=0)
mean_int_ring
```

#  graphical representation of pixels variation over time for ROI 2.

```python
kymograph = roi.kymograph(X_data, label_array1, 2)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(kymograph)
plt.title("pixels variation over time")
plt.ylabel("time")
plt.xlabel("pixels")
plt.ylim(300,400)
plt.colorbar()
plt.show()
```

## The brightest pixel in any ROI in any image in the image set

```python
max_cts = roi.roi_max_counts(X_data, label_array1)
max_cts
```

#  X-ray Spcekle visibility spectroscopy ( XSVS)

```python

spe_cts_all, std_dev = xsvs.xsvs((X_data,), label_array1, timebin_num=2,
                             number_of_img=50, max_cts=15)
```

```python
spe_cts_all.shape
```

#  Integartion times

```python
time_steps = core.geometric_series(2, 50)
time_steps
```

## the normalized bin edges and bin centers for each integration time.

```python
num_times = 6
num_rois=2
norm_bin_edges, norm_bin_centers = xsvs.normalize_bin_edges(num_times,
                                                            num_rois, mean_int_ring, max_cts=15)
```

```python
norm_bin_edges.shape
```

```python
spe_cts_all[3,0].shape, norm_bin_edges[3,0][:-1].shape
```

# 1st ROI

```python
fig, axes = plt.subplots(figsize=(6, 6))
axes.set_xlabel("K/<K>")
axes.set_ylabel("P(K)")
for i in range(3):
    art, = axes.plot(norm_bin_edges[i, 0][:-1], spe_cts_all[i, 0], '-o', label=str(time_steps[i])+" ms")
    axes.set_xlim(0, 4)
    axes.legend()
plt.title("1st roi")
plt.show()
```

# 2 nd ROI

```python
fig, axes = plt.subplots(figsize=(6, 6))
axes.set_xlabel("K/<K>")
axes.set_ylabel("P(K)")
for i in range(3):
    art, = axes.plot(norm_bin_edges[i, 1][:-1], spe_cts_all[i, 1], '-o', label=str(time_steps[i])+" ms")
    axes.set_xlim(0, 4)
    axes.legend()
plt.title("1st roi")
plt.show()
```
