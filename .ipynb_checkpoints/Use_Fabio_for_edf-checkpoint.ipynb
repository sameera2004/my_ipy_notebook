#Use Fabio to open images

```python
import fabio
import numpy as np


def open_edf(img_num, ext, dir_path):

    img = fabio.open(dir_path+str(img_num)+ext)
    img_dat = img.data
    return img_dat

data_dir = "/Volumes/Data/BeamLines/CHX/Luxi_description_files_for_duke/duke_cc_12C/"
ext = ".edf"
first_file = 1
last_file = 5001
img_array = []


#  open the images
for i in range(first_file, last_file):
    if (i<10):
        img_array.append(open_edf(i, ext, data_dir+"duke_cc_12c_000"))
    elif (9<i<100):
        img_array.append(open_edf(i, ext, data_dir+"duke_cc_12c_00"))
    elif (99<i<1000):
        img_array.append(open_edf(i, ext, data_dir+"duke_cc_12c_0"))
    else:
        img_array.append(open_edf(i, ext, data_dir+"duke_cc_12c_"))
        
```

```python
np.save(data_dir+"duke_img_1_5000.npy",np.asarray(img_array))

dark_img = 0

for i in range(5001, 5101):
    dark_img += (open_edf(i, ext, data_dir+"duke_cc_12c_"))

# save the normalized dark image in the data directory
np.save(data_dir+"duke_dark.npy", dark_img/100)


#  Get the dark images
duke_ndata = np.load(data_dir+"duke_img_1_5000.npy")
duke_dark = np.load(data_dir+"duke_dark.npy")

duke_data = []
for i in range(duke_ndata.shape[0]):
    duke_data.append(duke_ndata[i] - duke_dark)
    
duke_data = np.asarray(duke_data)   


## Get the average image of 5000 images
sum_img = np.average(duke_data, axis=0)        
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.imshow(sum_img, vmax=1e0,cmap='jet')
plt.title("Duke_cc_12c")
plt.show()
```
