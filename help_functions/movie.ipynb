#   Create a movie from set of images

```python
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt

dpi = 100

def ani_frame():
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    im = ax.imshow(get_data(0), cmap='jet', interpolation='nearest')
    #im.set_clim([0, 1])
    fig.set_size_inches([5,5])

    tight_layout()

    def update_img(n):
        #tmp = rand(300, 300)
        if n == 100:
            tmp = rand(150, 150)
        else:
            tmp = get_data(n)
        im.set_data(tmp)
        return im

    #legend(loc=0)
    ani = animation.FuncAnimation(fig, update_img, 400, interval=20)
    #writer = animation.writers['ffmpeg'](fps=30)
    writer = animation.writers['mencoder'](fps=15)#avconv

    ani.save('nipa.mp4', writer=writer, dpi=dpi)
    #ani.save('demo.mp4', writer = 'mencoder', fps=15, dpi=dpi)
    return ani


def get_data(n):
    data = np.load("100_500_n250.npy")
    return data[n]


def save_data():
    data_dir = "/Volumes/Data/BeamLines/CHX/Luxi_description_files_for_duke/duke_cc_12C/"
    duke_ndata = np.load(data_dir+"duke_img_1_5000.npy")
    duke_dark = np.load(data_dir+"duke_dark.npy")

    duke_data = []
    for i in range(duke_ndata.shape[0]):
        duke_data.append(duke_ndata[i] - duke_dark)

    duke_data = np.asarray(duke_data)
    np.save("duke_data.npy", duke_data)

def ani_frame():
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    im = ax.imshow(rand(300, 300), cmap='jet', interpolation='nearest')
    im.set_clim([0, 1])
    fig.set_size_inches([5, 5])

    tight_layout()

    def update_img(n):
        tmp = rand(300, 300)
        #tmp = get_data(n)
        im.set_data(tmp)
        return im

    #legend(loc=0)
    ani = animation.FuncAnimation(fig, update_img, 300, interval=30)
    #writer = animation.writers['ffmpeg'](fps=30)
    writer = animation.writers['mencoder'](fps=30)#avconv

    ani.save('demo.mp4', writer = 'mencoder', fps=15, dpi=dpi)
    return ani
```
