# Finding minma and maxima of a data sets

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import argrelextrema
```

```python
y = np.arange(0, 50)
y
```

```python
x = np.random.random(50)

plt.figure(figsize=(10, 10))
plt.plot(y,x)
plt.show()
```

## for local minima

```python
# for local minima
minma = argrelextrema(x, np.less)
```

```python
(y[minma], x[minma])
```

## for local minima

```python
# for local minima
maxma = argrelextrema(x, np.greater)
```

```python
(y[maxma], x[maxma])
```
