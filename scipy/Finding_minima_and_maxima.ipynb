```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import argrelextrema

x = np.random.random(50)


plt.figure(figsize=(10, 10))
plt.plot(x)
plt.show()
```

```python
# for local minima
argrelextrema(x, np.less)
```

```python
# for local minima
argrelextrema(x, np.greater)
```
