$ gamma.pdf(x, a) = \frac{lambda^a * x^{a-1} * exp(-lambda*x) }{\Gamma(a)}$

### Gamma distribution

$  P(K) =(\frac{M}{<K>})^M \frac{K^(M-1)}{\Gamma(M)}\exp(-M\frac{K}{<K>})$ 

- M = a is the conhernet modes,
- x = K
- lambda = M/$<K>$  and  $<K>$ is the mean intensity

```python
rv = stats.gamma(3., loc = 0., scale = 2.)
```

scale = 1.0/lambda

```python
a = 1.99323054838
x = np.linspace(stats.gamma.ppf(0.01, a), stats.gamma.ppf(0.99, a), 100)
x
```

```python
fig, ax = plt.subplots()
ax.plot(x, stats.gamma.pdf(x, a), 'r-', lw=5, alpha=0.6, label='gamma pdf')
plt.show()
```

```python
rv = stats.gamma(a)
fig, ax = plt.subplots()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
vals = stats.gamma.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], stats.gamma.cdf(vals, a))
```

```python
# plot the distributions
k_values = [1, 2, 3, 5]
theta_values = [2, 1, 1, 0.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(1E-6, 10, 1000)

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(5, 3.75))

for k, t, ls in zip(k_values, theta_values, linestyles):
    dist = stats.gamma(k, 0, t)
    plt.plot(x, dist.pdf(x), ls=ls, 
             label=r'$k=%.1f,\ \theta=%.1f$' % (k, t))

plt.xlim(0, 10)
plt.ylim(0, 0.45)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|k,\theta)$')
plt.title('Gamma Distribution')

plt.legend(loc=0)
plt.show()
```
