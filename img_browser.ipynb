# This will display images interactively

```python
import numpy as np
import os

from IPython.html.widgets import *
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.html.widgets import interact, IntSlider, FloatSliderWidget
```

```python
img_stack = np.random.randint(1, 5, size=(500, ) + (500, 500))
```

```python
def browse_tif_images(img_stack, sample_name):
    n = len(img_stack)
    def view_image(i):
        plt.imshow(img_stack[i], interpolation='nearest')
        plt.title(sample_name+str(i))
        plt.show()
    interact(view_image, i=(0,n-1))
```

```python

browse_tif_images(img_stack, "Random  ")
```
