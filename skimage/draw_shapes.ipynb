```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.draw import ellipse
import numpy as np

img = np.zeros((10, 10), dtype=np.uint8)
rr, cc = ellipse(5, 5, 3, 4)
img[rr, cc] = 1
img
```

```python
fig, ax = plt.subplots()
ax.imshow(img, cmap='viridis')
plt.show()
```

```python
from skimage.draw import ellipse_perimeter
img = np.zeros((10, 10), dtype=np.uint8)
rr, cc = ellipse_perimeter(5, 5, 3, 4)
img[rr, cc] = 1
img
```

```python
fig, ax = plt.subplots()
ax.imshow(img, cmap='viridis')
plt.show()
```

```python
from skimage.draw import polygon
img = np.zeros((20, 20), dtype=np.uint8)
x = np.array([1, 7, 4, 1])
y = np.array([1, 2, 8, 1])
rr, cc = polygon(y, x)
img[rr, cc] = 1
img
```

```python
x = np.array([3, 6, 4, 8])
y = np.array([0, 2, 8, 1])
rr, cc = polygon(y, x)
img[rr, cc] = 1
img
```
